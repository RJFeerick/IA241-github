# Analysis of 2020 Covid-19 Data for Australia 

## Background of Covid-19

COVID-19 is a disease caused by a virus named SARS-CoV-2 and was discovered in December 2019 in Wuhan, China. It is very contagious and has quickly spread around the world. The pandemic started in January 2020 for Australians and the website below will provide specfic Covid-19 data for Australia.

## Data Source

This data is from [European Centre for Disease and Control]('https://www.ecdc.europa.eu/en/publications-data/download-todays-data-geographic-distribution-covid-19-cases-worldwide')

## Import Data from S3 Bucket

```python
%matplotlib inline
import pandas
```

```python
df = pandas.read_excel('s3://ia241-1-spring2020-feerick/covid_data.xls')
df[:10]
```

```python
australia_data =df.loc[df['countriesAndTerritories']  ==  'Australia']
australia_data[:10]
```

## Question 1: How Many Covid-19 Cases Have Been Confirmed as of December 2020?

```python
australia_data.sum()['cases']
```

```python
sum_cases_per_day = australia_data.groupby('month').sum()['cases']
sum_cases_per_day.plot()
```

The data shows that there was 28,031 cases in Australia as of December 2020. From the graph above you can see a spike the cases in March and then again in July and August. Although Australia was in lockdown longer than any other country they were shocked to see the spike in July. Researchers could not figure out why this happened or what caused, scientist speculate that it could be cause by mutation of Covid making it more contagious.

## Question 2: How are the Number of Deaths Related to the Number of Cases?

```python
australia_data.plot.scatter(x='cases',y='deaths',c='month')
```

Regarding the data shown in the graph above, it shows that the more cases there were in the fewer deaths. On the far left bottom corner of the graph it depicts that there were many months that had low cases and low deaths. However, an outliar in the data suggests that around months 6-8 there were a spike in cases and deaths. One can see from this graph that around June to August there were a spike in Covid-19 deaths and cases in Australia.

## Question 3: Which Month of 2020 yielded the most deaths due to Covid-19 in Australia?

```python
sum_deaths_per_month = australia_data.groupby('month').sum()['deaths']
```

```python
sum_deaths_per_day = australia_data.groupby('month').sum()['deaths']
sum_deaths_per_day.plot.bar()
```

The bar graph shown about visualizes the deaths per month in 2020 due to Covid-19. The highest recording months are July, August and September, after this it falls down to the average throughout the year. The month of August is a heavy outliar in the data shown above. One could imply that hot climates could play a roll in Covid-19 death rates.
